## 6. Merge the imported geometry

```python
office = objects[0]
merged = Topology.SelfMerge(office)

print("After SelfMerge")
print("Faces:", len(Cluster.Faces(merged)))
print("Shells:", len(Cluster.Shells(merged)))
print("Cells:", len(Cluster.Cells(merged)))
```

## 7. Extract the shells from the merged geometry

```python
shells = Cluster.Shells(merged)

print("Number of shells:", len(shells))
```

## 8. Convert the shells into cells

```python
cells = []

for s in shells:
    c = Cell.ByShell(s)
    if c:
        cells.append(c)

print("Recovered cells:", len(cells))
```

## 9. Build a CellComplex from the cells

```python
cc = CellComplex.ByCells(cells)

print("CellComplex created")
```

## 10. Derive the adjacency graph (g01)

```python
g1 = Graph.ByTopology(
    cc,
    useInternalVertex=True
)

print("Adjacency graph created")
print("Vertices:", len(Graph.Vertices(g1)))
print("Edges:", len(Graph.Edges(g1)))
```

## 11. Assign visual attributes to the vertices and edges of the graph (g01)

```python
vertices = Graph.Vertices(g1)

for v in vertices:
    d = Dictionary.ByKeysValues(
        ["size", "color"],
        [18, "red"]
    )
    v = Topology.SetDictionary(v, d)

edges = Graph.Edges(g1)

for e in edges:
    d = Dictionary.ByKeysValues(
        ["width", "color"],
        [6, "black"]
    )
    e = Topology.SetDictionary(e, d)
```

## 12. Show the geometry and the graph (g01)

```python
Topology.Show(
    cc,
    g1,

    vertexSizeKey="size",
    vertexColorKey="color",
    edgeWidthKey="width",
    edgeColorKey="color",

    faceOpacity=0.15,
    faceColor=[220,220,245],

    edgeWidth=0.5,
    edgeColor="lightgray",

    backgroundColor="white",
    width=900,
    height=700,
    renderer=renderer
)
```

## 14. Import the door geometry from Rhino

```python
doors = Topology.ByOBJPath(r"C:\PROJECTS\GRAPH-ML-DOCUMENTS\HW01\RhinoGeometry-Doors-Nihan.obj")

print("Doors object imported")
print(doors)
```

## 15. Extract the door cluster

```python
door_cluster = doors[0]

print("Door Faces:", len(Cluster.Faces(door_cluster)))
print("Door Shells:", len(Cluster.Shells(door_cluster)))
print("Door Cells:", len(Cluster.Cells(door_cluster)))
```

## 16. Extract the door faces

```python
door_faces = Topology.Faces(door_cluster)

print("Number of door faces:", len(door_faces))
```

## 17. Add the door faces as apertures to the cell complex

```python
cc = Topology.AddApertures(cc, door_faces)
print("Apertures added to the cell complex")
```

## 18. Derive the access graph

```python
g2 = Graph.ByTopology(
    cc,
    direct=False,
    useInternalVertex=True,
    directApertures=True,
)
```

## 19. Assign visual attributes to the vertices and edges of the access graph(g02)

```python
vertices = Graph.Vertices(g2)

for v in vertices:
    d = Dictionary.ByKeysValues(
        ["size", "color"],
        [18, "red"]
    )
    v = Topology.SetDictionary(v, d)

edges = Graph.Edges(g2)

for e in edges:
    d = Dictionary.ByKeysValues(
        ["width", "color"],
        [8, "black"]
    )
    e = Topology.SetDictionary(e, d)


print("Access graph created")
print("Vertices:", len(Graph.Vertices(g2)))
print("Edges:", len(Graph.Edges(g2)))
```

## 20. Show the geometry and the access graph (g02)

## 21. Derive the access graph (g03)

```python
g3 = Graph.ByTopology(
    cc,
    direct=False,
    useInternalVertex=True,
    viaSharedApertures=True,
    toExteriorApertures=True,
)

print("Access graph created")
print("Vertices:", len(Graph.Vertices(g3)))
print("Edges:", len(Graph.Edges(g3)))
```

## 22. Assign visual attributes

```python
g1_positions = set()
for v in Graph.Vertices(g1):
    pos = (round(Vertex.X(v), 3), round(Vertex.Y(v), 3), round(Vertex.Z(v), 3))
    g1_positions.add(pos)

# coloring door vertices
vertices = Graph.Vertices(g3)


for v in vertices:
    pos = (round(Vertex.X(v), 3), round(Vertex.Y(v), 3), round(Vertex.Z(v), 3))
    
    if pos in g1_positions:
        color = "red"  
        size  = 18
    else:
        color = "cyan"  
        size  = 14
    
    v_dict = Dictionary.ByKeysValues(["size", "color"], [size, color])
    Topology.SetDictionary(v, v_dict)


edges = Graph.Edges(g3)

for e in edges:
    e_dict = Dictionary.ByKeysValues(
        ["width", "color"],
        [4, "black"]   
    )
    Topology.SetDictionary(e, e_dict)

print("Visual attributes assigned to g3")
print("Vertices:", len(Graph.Vertices(g3)))
print("Edges:", len(Graph.Edges(g3)))
```
