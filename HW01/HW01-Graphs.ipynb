## 6. Merge the imported geometry

```python
office = objects[0]
merged = Topology.SelfMerge(office)

print("After SelfMerge")
print("Faces:", len(Cluster.Faces(merged)))
print("Shells:", len(Cluster.Shells(merged)))
print("Cells:", len(Cluster.Cells(merged)))
```

## 7. Extract the shells from the merged geometry

```python
shells = Cluster.Shells(merged)

print("Number of shells:", len(shells))
```

## 8. Convert the shells into cells

```python
cells = []

for s in shells:
    c = Cell.ByShell(s)
    if c:
        cells.append(c)

print("Recovered cells:", len(cells))
```

## 9. Build a CellComplex from the cells

```python
cc = CellComplex.ByCells(cells)

print("CellComplex created")
```

## 10. Derive the adjacency graph

```python
g1 = Graph.ByTopology(cc)

print("Graph created")
print("Vertices:", len(Graph.Vertices(g1)))
print("Edges:", len(Graph.Edges(g1)))
```

## 11. Assign visual attributes to the vertices and edges of the graph

```python
vertices = Graph.Vertices(g1)

for v in vertices:
    d = Dictionary.ByKeysValues(
        ["size", "color"],
        [18, "red"]
    )
    v = Topology.SetDictionary(v, d)

edges = Graph.Edges(g1)

for e in edges:
    d = Dictionary.ByKeysValues(
        ["width", "color"],
        [4, "black"]
    )
    e = Topology.SetDictionary(e, d)
```

## 12. Show the geometry and the graph

```python
Topology.Show(
    cc,
    g1,
    vertexSizeKey="size",
    vertexColorKey="color",
    edgeWidthKey="width",
    edgeColorKey="color",
    faceOpacity=0.3,
    edgeWidth=1,
    edgeColor="black",
    backgroundColor="white",
    width=900,
    height=700,
    renderer=renderer
)
```
