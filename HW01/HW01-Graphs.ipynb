## 6. Merge the imported geometry

```python
office = objects[0]
merged = Topology.SelfMerge(office)

print("After SelfMerge")
print("Faces:", len(Cluster.Faces(merged)))
print("Shells:", len(Cluster.Shells(merged)))
print("Cells:", len(Cluster.Cells(merged)))
```

## 7. Extract the shells from the merged geometry

```python
shells = Cluster.Shells(merged)

print("Number of shells:", len(shells))
```

## 8. Convert the shells into cells

```python
cells = []

for s in shells:
    c = Cell.ByShell(s)
    if c:
        cells.append(c)

print("Recovered cells:", len(cells))
```

## 9. Build a CellComplex from the cells

```python
cc = CellComplex.ByCells(cells)

print("CellComplex created")
```

## 10. Derive the adjacency graph

```python
g1 = Graph.ByTopology(
    cc,
    useInternalVertex=True
)

print("Adjacency graph created")
print("Vertices:", len(Graph.Vertices(g1)))
print("Edges:", len(Graph.Edges(g1)))
```

## 11. Assign visual attributes to the vertices and edges of the graph

```python
vertices = Graph.Vertices(g1)

for v in vertices:
    d = Dictionary.ByKeysValues(
        ["size", "color"],
        [18, "red"]
    )
    v = Topology.SetDictionary(v, d)

edges = Graph.Edges(g1)

for e in edges:
    d = Dictionary.ByKeysValues(
        ["width", "color"],
        [6, "black"]
    )
    e = Topology.SetDictionary(e, d)
```

## 12. Show the geometry and the graph

```python
Topology.Show(
    cc,
    g1,

    vertexSizeKey="size",
    vertexColorKey="color",
    edgeWidthKey="width",
    edgeColorKey="color",

    faceOpacity=0.15,
    faceColor=[220,220,245],

    edgeWidth=0.5,
    edgeColor="lightgray",

    backgroundColor="white",
    width=900,
    height=700,
    renderer=renderer
)
```

## 14. Import the door geometry from Rhino

## 15. Extract the door cluster

## 16. Extract the door faces

```python
g2 = Graph.ByTopology(
    cc,
    direct=False,
    directApertures=True,
    useInternalVertex=True
)
```

## 21. Extract the faces of the cell complex

```python
cc_faces = CellComplex.Faces(cc)

print("Number of cell complex faces:", len(cc_faces))
```

## 22. Find the nearest host face for each door face

```python
host_faces = []

for df in door_faces:
    door_center = Topology.Centroid(df)

    min_dist = 1e9
    nearest_face = None

    for cf in cc_faces:
        face_center = Topology.Centroid(cf)
        dist = Vertex.Distance(door_center, face_center)

        if dist < min_dist:
            min_dist = dist
            nearest_face = cf

    host_faces.append(nearest_face)

print("Host faces found:", len(host_faces))
```

## 23. Add each door face as an aperture to its host face


```python
new_faces = []

for i in range(len(door_faces)):
    df = door_faces[i]
    hf = host_faces[i]

    new_face = Topology.AddApertures(hf, [df])
    new_faces.append(new_face)

print("Apertures assigned to host faces")
```

## 24. Check which faces now contain apertures

```python
faces_with_apertures = []

for f in new_faces:
    aps = Topology.Apertures(f)
    if aps and len(aps) > 0:
        faces_with_apertures.append(f)

print("Faces with apertures:", len(faces_with_apertures))
```

## 25. Extract only the vertical faces of the cell complex

```python
from topologicpy.Face import Face

vertical_faces = []

for f in cc_faces:
    n = Face.Normal(f)
    
    # normalin z bileşeni küçükse yüzey dikey kabul edilir
    if abs(n[2]) < 0.2:
        vertical_faces.append(f)

print("Vertical faces:", len(vertical_faces))
```

## 26. Find the nearest vertical host face for each door face


```python
host_faces = []

for df in door_faces:
    door_center = Topology.Centroid(df)

    min_dist = 1e9
    nearest_face = None

    for vf in vertical_faces:
        face_center = Topology.Centroid(vf)
        dist = Vertex.Distance(door_center, face_center)

        if dist < min_dist:
            min_dist = dist
            nearest_face = vf

    host_faces.append(nearest_face)

print("Host faces found:", len(host_faces))
```

## 27. Add each door face as an aperture to its vertical host face

```python
new_faces = []

for i in range(len(door_faces)):
    df = door_faces[i]
    hf = host_faces[i]

    new_face = Topology.AddApertures(hf, [df])
    new_faces.append(new_face)

print("Apertures assigned to vertical host faces")
```

## 28. Check which faces now contain apertures

```python
faces_with_apertures = []

for f in new_faces:
    aps = Topology.Apertures(f)
    if aps and len(aps) > 0:
        faces_with_apertures.append(f)

print("Faces with apertures:", len(faces_with_apertures))
```

## 29. Derive the access graph

```python
g2 = Graph.ByTopology(
    cc,
    direct=False,
    directApertures=True,
    useInternalVertex=True
)

print("Access graph created")
print("Vertices:", len(Graph.Vertices(g2)))
print("Edges:", len(Graph.Edges(g2)))
```

## 30. Assign visual attributes
